```python
import pandas as pd
import matplotlib.pyplot as plt

# Load benchmarking results
results_path = '../reports/model_benchmark_results.csv'
results_df = pd.read_csv(results_path)

# Sort models by accuracy
results_df = results_df.sort_values(by='Accuracy', ascending=False)
print("Benchmarking Results:\n", results_df)

# Plot results for visualization
plt.figure(figsize=(10, 6))
plt.barh(results_df['Model'], results_df['Accuracy'], color='skyblue')
plt.xlabel('Accuracy')
plt.title('Model Benchmarking Results')
plt.gca().invert_yaxis()  # Invert Y-axis for better readability
plt.savefig('../reports/model_selection_visualization.png')
plt.show()

# Select the top models
top_models = results_df.head(2)['Model'].values
print("\nTop Models Selected for Further Steps:")
print("\n".join(top_models))

# Save selected models
selected_models_path = '../reports/selected_models.txt'
with open(selected_models_path, 'w') as f:
    f.write('\n'.join(top_models))

print(f"\nSelected models saved to: {selected_models_path}")
```
